```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data.head()
```

```python
# Fix train data
n_train_data = train_data.select_dtypes(include=[np.number])
n_train_data.head()

print(n_train_data.isnull().sum())

# Median age missing data
n_train_data['Age'] = n_train_data['Age'].fillna(n_train_data['Age'].median())


print(n_train_data.isnull().sum())
```

```python
# Fix test data
n_test_data = test_data.select_dtypes(include=[np.number])
print(n_test_data.isnull().sum())
n_test_data['Age'] = n_test_data['Age'].fillna(n_test_data['Age'].median())

#n_test_data.head()

n_test_data['Fare'] = n_test_data['Fare'].fillna(n_test_data['Fare'].median())

print(n_test_data.isnull().sum())
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Data was logged, not pre-processing needed.

# Define features and target
x = n_train_data.drop('Survived', axis=1)
y = n_train_data['Survived']

# Test split was already done, need to do validation split.

# x_train, x_val, y_train, y_val = train_test_split(
#     x,y,
#     test_size=0.25,
#     random_state=42
# )

log = LogisticRegression(
    random_state=42, max_iter=1000
)
log.fit(x, y)

y_pred = log.predict(n_test_data)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': y_pred})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
